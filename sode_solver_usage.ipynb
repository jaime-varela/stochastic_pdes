```julia
d = 2; m = 2;
A = [1 0; 0 -1]
sigma = [1, 2]
T = 1; N = 5000; u0 = [1, 1]

t, u = MilsteinDiag(u0, T, N, d, m,
    u -> A * u,
    u -> sigma .* u,
    u -> sigma)

plot(t, u[1, :], label="u[1]")
plot!(t, u[2, :], label="u[2]")
xlabel!("t")
ylabel!("u")
title!("Fig 8.6(a)")
```
